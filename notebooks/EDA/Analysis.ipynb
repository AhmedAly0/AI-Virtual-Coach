**Imports**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

# Load the uploaded CSV file
file_path = "metadata - Sheet1.csv"
df = pd.read_csv(file_path)

# Display basic info and first few rows of the dataset
df.info(), df.head()
```

```python
# Create a BMI column
df["BMI"] = df["Weight"] / ((df["Height"] / 100) ** 2)

# Set up seaborn style
sns.set(style="whitegrid")

# 1. Gender Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender ', palette="pastel")
plt.title("Gender Distribution")
plt.tight_layout()
plt.show()

# 2. Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', bins=10, kde=True, color="skyblue")
plt.title("Age Distribution")
plt.tight_layout()
plt.show()

# 3. Weight Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Weight', bins=10, kde=True, color="salmon")
plt.title("Weight Distribution")
plt.tight_layout()
plt.show()

# 4. Height Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Height', bins=10, kde=True, color="lightgreen")
plt.title("Height Distribution")
plt.tight_layout()
plt.show()

# 5. BMI Distribution
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='BMI', bins=10, kde=True, color="mediumpurple")
plt.title("BMI Distribution")
plt.tight_layout()
plt.show()
```
